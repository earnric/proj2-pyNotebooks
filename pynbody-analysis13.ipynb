## Plots

```python
(len(s.g['pzf'][s.g['pzf']>0.0]),s.g['pzf'][s.g['pzf']>0.0])
```

```python
#Ignore ridiculously low primordial Z in the gas... 
s.g['pzf'][s.g['pzf']<1e-20] = 0.0
```

```python
(len(s.g['pzf'][s.g['pzf']>0.0]),s.g['pzf'][s.g['pzf']>0.0],max(s.g['pzf']))
```

```python
titleStr = "Primordial Z fraction @ z = %2d" % z + "\nVol ave"
sph.image(s.g,qty="pzf",width=boxsizestring,cmap="YlOrRd", denoise=True ,av_z=True, vmin=1e-20, vmax=0.3,
              approximate_fast=False,title=titleStr); 
titleStr = "Z @ z = %2d" % z + "\nVol ave"
sph.image(s.g,qty="metal",width=boxsizestring,cmap="YlOrRd", denoise=True ,av_z=True, log=False,
              approximate_fast=False,title=titleStr); #vmin=0.006, vmax=1.0,
```

```python
(s.g['pgf'][s.g['pgf']<1.0],min(s.g['pgf']))
```

```python
titleStr = "Z @ z = %2d" % z + "\nVol ave"
sph.image(s.g,qty="metal",width=boxsizestring,cmap="YlOrRd", denoise=True ,av_z=True, log=False,
              approximate_fast=False,title=titleStr); #vmin=0.006, vmax=1.0,
```

```python
titleStr = "Pristine gas fraction @ z = %2d" % z + "\nVol ave"
sph.image(s.g,qty="pgf",width=boxsizestring,cmap="YlOrRd", denoise=True ,av_z=True, log=False,
              approximate_fast=False,title=titleStr); #vmin=0.006, vmax=1.0,
```

```python
#Plotting a slice at z=0
titleStr = "Mass density @ z = %2d" % z + "\nslice @ zCoord = 0"
pynbody.plot.image(s.g,width=boxsizestring, units='Msol kpc^-2', log=False, center=True, cmap=cm.jet,title=titleStr);
```

```python
with pynbody.transformation.translate(s,[0,0,5]):
    titleStr = "Mass density @ z = %2d" % z + "\nslice @ zCoord = 5"
    pynbody.plot.image(s.g,width=boxsizestring, center=True, cmap=cm.jet,title=titleStr);
```

```python
    titleStr = "Mass density @ z = %2d" % z + "\nVol Ave"
    pynbody.plot.image(s.g,width=boxsizestring, log=False , av_z=True, center=True, cmap=cm.jet,title=titleStr);
```

## Plot temperature data

```python
titleStr = "Gas Temp @ z = %2d" % z + "\nVol ave"
sph.image(s.g,qty="temp",width=boxsizestring,cmap="YlOrRd", denoise=True ,av_z=True, 
              approximate_fast=False,title=titleStr); #vmin=0.1, vmax=2.0e4,
```

```python
with pynbody.transformation.translate(s,[0,0,5]):
    titleStr = "Gas Temp @ z = %2d" % z + "\nslice @ zCoord = 5"
    sph.image(s.g,qty="temp",width=boxsizestring,cmap="YlOrRd",  denoise=True ,av_z=False,log=False,
              approximate_fast=False,title=titleStr); #vmin=0.1, vmax=2.0e4,
```

```python
titleStr = "Gas Temp @ z = %2d" % z + "\nVol Weighted Ave thru box"
sph.image(s.g,qty="temp",width=boxsizestring,cmap="YlOrRd", denoise=True ,av_z=True,
          approximate_fast=False,title=titleStr);#vmin=100, vmax=5.0e4,
```

```python
titleStr = "Gas Temp @ z = %2d" % z + "\nDensity Weighted Ave thru box"
sph.image(s.g,qty="temp",width=boxsizestring,cmap="YlOrRd", denoise=True ,av_z="rho", vmin=0.1, vmax=2.0e4,
          approximate_fast=False,title=titleStr);
```
