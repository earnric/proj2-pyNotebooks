```python
z=ds.current_redshift
print "Redshift ", z
```

```python
p = yt.ProjectionPlot(ds, "z", ["density", "temperature"], weight_field="density")
p.show()
```

#### And, we can re-center the plot on different locations.  One possible use of this would be to make a single `ProjectionPlot` which you move around to look at different regions in your simulation, saving at each one.

```python
v, c = ds.find_max("density")
```

```python
v
```

```python
c
```

##### Center on the x-y coord, since the default looks down z

```python
p.set_center((c[0], c[1]))
p.zoom(10)
```

```python
my_sphere = ds.sphere("c", wid)
```

```python
# fractional=True keyword -- divide the profile data by the sum. 
plot = yt.PhasePlot(my_sphere, "density", "temperature", "cell_mass", weight_field=None, fractional=True )
plot.show()
```

```python
plot2 = yt.PhasePlot(my_sphere, "density", "temperature", "cell_mass", weight_field=None, fractional=False )
plot2.show()
```

```python
plot3 = yt.PhasePlot(my_sphere, "density", "temperature", "cell_mass", weight_field="temperature", fractional=True )
plot3.show()
```

```python
plot.set_unit("density","Msun/pc**3")
```
