```python
import pynbody
import pynbody.plot.sph as sph
from mpl_toolkits.axes_grid1 import ImageGrid
pynbody.ramses.multiprocess_num = 256
pynbody.config['number_of_threads'] = 8
```

```python
sc = pynbody.load('/Users/earnric/RAMSES-Data/fixedStandardCool/output_00009')
ec = pynbody.load('/Users/earnric/RAMSES-Data/fixedEnhancedCool/output_00010')
```

### Actually load the data. This is done when you ask for the data.

```python
sc['pos'];ec['pos'];
```

```python
sc.properties,ec.properties
```

```python
zs=1/sc.properties['a']-1
ze=1/ec.properties['a']-1
print ("Standard cool z= %.2f" % zs)
print ("Enhanced cool z= %.2f" % ze)
```

```python
boxsizestrings = "%.2f" % sc.properties['boxsize'].in_units('kpc')
boxsizestrings += " kpc"
boxsizestringe = "%.2f" % ec.properties['boxsize'].in_units('kpc')
boxsizestringe += " kpc"
print ("Physical sim-size at this z is %s"%boxsizestrings)
print ("Physical sim-size at this z is %s"%boxsizestringe)
```

# NORMALIZE the positions of the 'particles' (cells) in the sim. They are stored by ramses with origin 0,0,0 and the box has dimension 1,1,1

```python
sc['pos'] -= 0.5
ec['pos'] -= 0.5
```

```python
sc.physical_units()
ec.physical_units()
```

### Dark matter distribution - standard cooling

```python
titleStr = "Standard cooling model @ z = %.1f\nDark Matter Mass - Vol ave"%zs
pynbody.plot.image(sc.d,width=boxsizestrings, units='Msol kpc^-2', cmap=cm.Greys, title=titleStr);
```

```python
titleStr = "Enhanced cooling model @ z = %.1f\nDark Matter Mass - Vol ave"%ze
pynbody.plot.image(ec.d,width=boxsizestringe, units='Msol kpc^-2', cmap=cm.Greys, title=titleStr);
```

## Look at the Pristine Gas Fraction scalar for the gas

### Normalize metals - truncate small values to lower than 10^-5

#### Convert metallicity solar units

```python
sc.gas['metal'] *= 50
ec.gas['metal'] *= 50
```

```python
# Truncate to critical metallicity
sc.gas['metal'][sc.gas['metal']< 1.0e-6]=1.0e-6
ec.gas['metal'][ec.gas['metal']< 1.0e-6]=1.0e-6
```

```python
sc.gas['pgf'][sc.gas['pgf']<1.0e-6]=1.0e-6
ec.gas['pgf'][ec.gas['pgf']<1.0e-6]=1.0e-6
```

```python
i=7
```

```python
with pynbody.transformation.translate(ec,[0,0,i]):
    titleStr = "Standard cool z = %.2f\nZ - Slice @ $z_{coord}$ = %i"%(zs,i)
    titleStr = "Standard cool z = %.2f\nZ - Vol Ave"%(zs)
    sph.image(sc.g,qty="metal",width=boxsizestringe,cmap="YlOrRd", av_z=True, denoise=False ,qtytitle ='$log_{10}\, Z_{\odot}$',
              approximate_fast=False,title=titleStr);
```

```python
with pynbody.transformation.translate(ec,[0,0,i]):
    titleStr = "Enhanced cool z = %.2f\nZ - Slice @ $z_{coord}$ = %i"%(ze,i)
    titleStr = "Enhanced cool z = %.2f\nZ - Vol Ave"%(ze)
    sph.image(ec.g,qty="metal",width=boxsizestringe,cmap="YlOrRd", av_z=False, denoise=False ,qtytitle ='$log_{10}\, Z_{\odot}$',
              approximate_fast=False,title=titleStr);
```
