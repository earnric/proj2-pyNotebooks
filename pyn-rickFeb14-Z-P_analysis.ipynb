```python
import pynbody
import pynbody.plot.sph as sph
import mmap
pynbody.ramses.multiprocess_num = 16
pynbody.config['number_of_threads'] = 64
```

```python
import matplotlib.pyplot as plt 
```

```python
s = pynbody.load('output_00004')
s['pos']
s['pos'] -= 0.5
s.physical_units();
```

```python
z = 1/s.properties['a']-1
print "Redshift =",z
boxsizestring = "%.2f" % s.properties['boxsize'].in_units('kpc')
boxsizestring += " kpc"
print boxsizestring
```

```python
s.g['metal'][s.g['metal']<1e-6] = 0.0
s.g['pzf'][s.g['pzf']<1e-6] = 0.0
s.s['ppf'][s.s['ppf']>1.0] = 1.0
linex=(1e-7,1)
```

```python
plt.loglog(s.g['metal'],1.0-s.g['pgf']+1e-6,'.',linex,linex)
plt.title("Z vs (1-Pf) z=%.1f"%z)
plt.grid(True, which="both")
axes().set_xlabel("Z")
axes().set_ylabel("1-P")
plt.ylim(1e-7, 1)
plt.xlim(1e-7, 1)
```
