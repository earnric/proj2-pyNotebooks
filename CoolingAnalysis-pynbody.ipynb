## Plots

```python
#Plotting a slice at z=0
titleStr = "Mass density @ z = %2d" % z + "\nslice @ zCoord = 0"
pynbody.plot.image(s.g,width=boxsizestring, units='Msol kpc^-3', center=True, cmap=cm.jet,title=titleStr);
```

```python
with pynbody.transformation.translate(s,[0,0,5]):
    titleStr = "Mass density @ z = %2d" % z + "\nslice @ zCoord = 5"
    pynbody.plot.image(s.g,width=boxsizestring, center=True, cmap=cm.jet,title=titleStr);
```
