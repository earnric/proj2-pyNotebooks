## Plot the metals in a slice at the center of the simulation box

#### Remember - you get a slice at box-center unless you ask for something else... 

```python
boxsizestring
```

```python
titleStr = "Z @ z = %.1f" % z + "\nVol weighted Ave"
sph.image(s.g,qty="metal",width=boxsizestring,cmap="YlOrRd",  av_z=True, denoise=False ,
          approximate_fast=False,title=titleStr);
```

```python
boxsizestring
```

```python
for i in [-35,-30,-25]:
    with pynbody.transformation.translate(s,[0,0,i]):
        titleStr = "Z @ z = %.1f" % z + "\nSlice @ zAxis %3d kpc" %i
        sph.image(s.g,qty="metal",width=boxsizestring,cmap="YlOrRd", av_z=False, denoise=False ,qtytitle="${log_{10} Z}$"
                  ,approximate_fast=False,title=titleStr)
        plt.show()
```

### Look at the Pristine Gas Fraction scalar for the gas

```python
for i in [-35,-30,-25]:
    with pynbody.transformation.translate(s,[0,0,i]):
        titleStr = "Pristing gas fraction @ z = %.1f" % z + "\nSlice @ zAxis %3d kpc"%i
        sph.image(s.g,qty="pgf",width=boxsizestring,cmap="YlOrRd", av_z=False, denoise=False ,qtytitle="${log_{10} P_f}$",
              approximate_fast=False,title=titleStr)
        plt.show()
```

```python
# create the subplots
f, axs = plt.subplots(1,2,figsize=(14,6))
i=-30
#create a simple slice showing the gas temperature, with velocity vectors overlaid
titleStr = "Pristing gas fraction @ z = %.1f" % z + "\nSlice @ zAxis %3d kpc"%i
with pynbody.transformation.translate(s,[0,0,i]):
    sph.image(s.g,qty="pgf",width=boxsizestring,cmap="YlOrRd", av_z=False, denoise=False ,qtytitle="${log_{10} P_f}$",
            approximate_fast=False,title=titleStr,subplot=axs[0],show_cbar=False)
    titleStr = "Z @ z = %.1f" % z + "\nSlice @ zAxis %3d kpc"%i
    sph.image(s.g,qty="metal",width=boxsizestring,cmap="YlOrRd", av_z=False, denoise=False ,qtytitle="${log_{10} Z}$",
            approximate_fast=False,title=titleStr,subplot=axs[1],show_cbar=False)
```

```python
# create the subplots
f, axs = plt.subplots(1,2,figsize=(14,6))
i=-30
#create a simple slice showing the gas temperature, with velocity vectors overlaid
titleStr = "Pristing gas fraction @ z = %.1f" % z + "\nSlice @ zAxis %3d kpc"%i
with pynbody.transformation.translate(s,[40,85,i]):
    sph.image(s.g,qty="pgf",width='50 kpc',cmap="YlOrRd", av_z=False, denoise=False ,qtytitle="${log_{10} P_f}$",
            approximate_fast=False,title=titleStr,subplot=axs[0],show_cbar=False)
    titleStr = "Z @ z = %.1f" % z + "\nSlice @ zAxis %3d kpc"%i
    sph.image(s.g,qty="metal",width='50 kpc',cmap="YlOrRd", av_z=False, denoise=False ,qtytitle="${log_{10} Z}$",
            approximate_fast=False,title=titleStr,subplot=axs[1],show_cbar=False)
```

```python
# create the subplots
f, axs = plt.subplots(1,2,figsize=(16,6))
i=-30
#create a simple slice showing the gas temperature, with velocity vectors overlaid
titleStr = "Pristing gas fraction @ z = %.1f" % z + "\nSlice @ zAxis %3d kpc"%i
with pynbody.transformation.translate(s,[40,85,i]):
    im1 = sph.image(s.g,qty="pgf",width='50 kpc',cmap="YlOrRd", av_z=False, denoise=False ,qtytitle="${log_{10} P_f}$",
            approximate_fast=False,title=titleStr,subplot=axs[0],show_cbar=False)
    titleStr = "Z @ z = %.1f" % z + "\nSlice @ zAxis %3d kpc"%i
    im2 = sph.image(s.g,qty="metal",width='50 kpc',cmap="YlOrRd", av_z=False, denoise=False ,qtytitle="${log_{10} Z}$",
            approximate_fast=False,title=titleStr,subplot=axs[1],show_cbar=False)
    cax = f.add_axes([0.9, 0.1, 0.03, 0.8])
    f.colorbar(im1, cax=cax)
    cax = f.add_axes([0.5, 0.1, 0.03, 0.8])
    f.colorbar(im2, cax=cax)
```
