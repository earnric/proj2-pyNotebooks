## Plot the metals in a slice at the center of the simulation box

#### Remember - you get a slice at box-center unless you ask for something else... 

```python
z = 1/s.properties['a']-1
titleStr = "Z @ z = %2d" % z
```

```python
sph.image(s.g,qty="metal",width='60.1 kpc',cmap="YlOrRd", denoise=False ,approximate_fast=False,title=titleStr);
```

```python
with pynbody.transformation.translate(s,[0,0,2]):
    titleStr = "Z @ z = %2d" % z + " Slice @ zcoord +2 kpc" 
    sph.image(s.g,qty="metal",width='60.1 kpc',cmap="YlOrRd", av_z=False, denoise=False ,approximate_fast=False,title=titleStr);
```

### Plot the average Z through the sim box

```python
titleStr = "Z @ z = %2d" % z + " averaged thru zcoord" 
sph.image(s.g,qty="metal",width='60.1 kpc',cmap="YlOrRd", av_z=True, qtytitle="log10 Z",
          denoise=False ,approximate_fast=False,title=titleStr);
```

```python
s.gas['pgf'][s.gas['pgf']<1e-10]=1e-10
```

```python
titleStr = "Pristine Gas Fraction @ z = %2d" % z + " averaged thru zcoord" 
sph.image(s.g,qty="pgf",width='60.1 kpc',cmap="YlOrRd", av_z=True, qtytitle="log10 Pf", 
          denoise=False ,approximate_fast=False,title=titleStr);
```
