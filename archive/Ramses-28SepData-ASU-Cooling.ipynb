## Look at the Pristine Gas Fraction scalar for the gas

### First check for items that are larger than 1.0... Select out the items that are < 1.0 in the array. 

```python
s.g['pgf'][s.g['pgf']>1.0].size
```

```python
s.g['pgf'][s.g['pgf']>1.0]=1.0
```

## Zero-out metals lower than 10^-10

```python
s.gas['metal'][s.gas['metal']<1e-10]>0].size
```

```python
s.gas['metal'][s.gas['metal']>1e-5].size
```

```python
s.gas['metal'][s.gas['metal']<1e-10]=1e-10
```

## Plot the metals in a slice at the center of the simulation box

#### Remember - you get a slice at box-center unless you ask for something else... 

```python
sph.image(s.g,qty="metal",width="32 kpc",cmap="YlOrRd", denoise=False ,approximate_fast=False,title="Z");
```

```python
with pynbody.transformation.translate(s,[0,0,2]):
    sph.image(s.g,qty="metal",width='384.1 kpc',cmap="YlOrRd", av_z=False, denoise=False ,approximate_fast=False,title="Z");
```

### Plot the average Z through the sim box

```python
sph.image(s.g,qty="metal",width='384.1 kpc',cmap="YlOrRd", av_z=True, qtytitle="log10 Z",denoise=False ,approximate_fast=False,title="Z");
```

```python
s.gas['pgf'][s.gas['pgf']<1e-10]=1e-10
```
