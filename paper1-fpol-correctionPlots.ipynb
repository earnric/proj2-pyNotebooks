```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
import numpy as np
import math as math
from IPython import display
import gc
plt.rcParams['figure.figsize'] = (10,8)
plt.rcParams['font.size'] = 20
```

```python
from matplotlib.ticker import FormatStrFormatter
from matplotlib.ticker import FuncFormatter
from matplotlib.ticker import NullFormatter

def log_formatter(x,pos):
    return format(np.log10(x), '.0f')

def div_formatter(x,pos):
    return format(x/1000, '.1f')

def GHzFormatter(x,pos):
    return format(x/1e9, '.2e')
GHzfmt  = FuncFormatter(GHzFormatter)
divfmt  = FuncFormatter(div_formatter)
logfmt  = FuncFormatter(log_formatter)
nullfmt = NullFormatter()
```

```python
Zcrit = 1e-5
zrange = np.logspace(-5,-2,100) # 1e-5, 1e-1
def upperCor(P):
    return np.array(zrange/(1-P))

def lowerCor(P):
    return np.array((zrange - Zcrit*P)/(1-P))
```

```python
Zcrit = 1e-5

redLine  = mlines.Line2D([], [], color='red', label=r'$Z=\frac{\overline {Z}}{(1-P)}$')
blueLine = mlines.Line2D([], [], color='blue',ls = '--', label=r'$Z=\frac{\overline {Z} - Z_{crit} P}{(1-P)}$')

fig = plt.figure(figsize=(12,8))
ax = plt.axes()
ax.set_xscale('log')
ax.set_yscale('log')

ax.plot(zrange,upperCor(0.0),'g')
ax.annotate('P=0.0', xy=(1e-4, 1e-2), xytext=(5e-4, 3.5e-4),
                  horizontalalignment='center', size=20,rotation=18.0 ) 

ax.plot(zrange,upperCor(0.5),'r-')
ax.plot(zrange,lowerCor(0.5),'b--')
ax.annotate('P=0.5', xy=(1e-4, 1e-2), xytext=(1e-4, 3e-4),
                  horizontalalignment='center', size=20,rotation=18.0 ) 

ax.plot(zrange,upperCor(0.9),'r-')
ax.plot(zrange,lowerCor(0.9),'b--')
ax.annotate('P=0.9', xy=(1e-4, 1e-2), xytext=(1e-4, 1.7e-3),
                  horizontalalignment='center', size=20,rotation=18.0 ) 

ax.plot(zrange,upperCor(0.995),'r-')
ax.plot(zrange,lowerCor(0.995),'b--')
ax.annotate('P=0.995', xy=(1e-4, 1e-2), xytext=(1e-4, 3.5e-2),
                  horizontalalignment='center', size=20,rotation=18.0 ) 

ax.plot(zrange,upperCor(0.999),'r-')
ax.plot(zrange,lowerCor(0.999),'b--')
ax.annotate('P=0.999', xy=(1e-4, 1e-2), xytext=(1e-4, 1.7e-1),
                  horizontalalignment='center', size=20,rotation=18.0 ) 

ax.xaxis.set_major_formatter(logfmt)
ax.yaxis.set_major_formatter(logfmt)

ax.grid()
# ax.axes.set_ylim([0.85,2.2])
# ax.axes.set_xlim([0.0,8.0])
ax.set_ylabel(r"${\rm log}\; Z$")
ax.set_xlabel(r'${\rm log}\; \overline{Z}$')

leg1 = ax.legend(handles=[redLine, blueLine],loc=2, fontsize=22)
# Add the legend manually to the current Axes.
pltAx = plt.gca().add_artist(leg1)

plt.savefig("fpolBounds.pdf",bbox_inches='tight')
```
