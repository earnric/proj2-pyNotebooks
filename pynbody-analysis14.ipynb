```python
import matplotlib.pyplot as plt 
```

```python
titleStr = "GAS PHASE @ z = %.1f" % z
pynbody.plot.gas.rho_T(s.g,rho_units='g cm**-3', rho_range=(-28.,-23.5), t_range=(-1,5),
                       colorbar=True, scalemin=0.1, scalemax=8.0e5, title=titleStr);
plt.title(titleStr)
plt.show()
```

```python
bs=s.properties['boxsize'].in_units('kpc')
```

```python
plt.plot(s.s['x'],s.s['y'],'.')
x1,x2,y1,y2 = plt.axis()
plt.axis((-bs/2.0,bs/2.0,-bs/2.0,bs/2.0))
plt.title("Star particle locations z=%.1f\n2D projection, xy plane, %s boxsize"%(z,boxsizestring))
```

```python
pynbody.plot.stars.sfh(s);
```

```python
# Ignore tiiny variations...
s.g['metal'][s.g['metal']<1e-10] = 0.0
s.g['pzf'][s.g['pzf']<1e-10] = 0.0
s.g['pzf'][s.g['pzf']>1.0] = 1.0
s.g['pgf'][s.g['pgf']<1e-10] = 0.0
s.g['pgf'][s.g['pgf']>1.0] = 1.0
```

```python
titleStr = "Primordial Z fraction @ z = %.1lf" % z + "\nVol ave"
sph.image(s.g,qty="pzf",width=boxsizestring,cmap="YlOrRd", denoise=True ,av_z=True, vmin=1e-10, 
              approximate_fast=False,title=titleStr); 
```

```python
titleStr = "Z @ z = %.1lf" % z + "\nVol ave"
sph.image(s.g,qty="metal",width=boxsizestring,cmap="YlOrRd", denoise=True ,av_z=True, log=True, vmin=1e-10,
              approximate_fast=False,title=titleStr); #vmin=0.006, vmax=1.0,
```

```python
titleStr = "Pristine gas fraction @ z = %.1lf" % z + "\nVol ave"
sph.image(s.g,qty="pgf",width=boxsizestring,cmap="YlOrRd", denoise=True ,av_z=True, log=True,
              approximate_fast=False,title=titleStr); #vmin=0.006, vmax=1.0,
```

```python
titleStr = "Pristine gas fraction @ z = %.1lf" % z + "\nDensity averaged"
sph.image(s.g,qty="pgf",width=boxsizestring,cmap="YlOrRd", denoise=True ,av_z="rho", log=True,
              approximate_fast=False,title=titleStr); #vmin=0.006, vmax=1.0,
```
